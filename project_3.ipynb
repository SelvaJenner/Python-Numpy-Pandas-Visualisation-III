10) Is there any difference b/w the votes of restaurants accepting and not accepting online orders?

```python
df.pivot_table( index =df['online_order'],values = ['votes'],aggfunc = ['sum','mean'])
```

```python
# 11) Which are the most common restaurant type in Banglore?
df['rest_type'].value_counts().iloc[0:10].plot(kind = 'bar', legend = True, rot = 45)
plt.show()
```

```python
# 12) Find the Best budget Restaurants in any location.
budget = df[(df.Approx_cost < df.Approx_cost.mean()) & (df.rate >= 4.0)]
budget.reset_index(drop = True)
```

```python
df[(df.Approx_cost < df.Approx_cost.mean()) & (df.rate >= 4.0)].sort_values('rate',ignore_index = True, ascending = False).iloc[0:10]
```

```python
# 13) Top quick bites restaurant chains in Banglore.

df[(df.rest_type == 'Quick Bites') & (df.rate >= 4.0)& (df.votes >=df.votes.mean())].sort_values('rate',ascending= False,ignore_index= True).iloc[0:10]
```
